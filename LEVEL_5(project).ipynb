#Project 1: Personal Budget Tracker

1. Open a Python editor.
2. Create a dictionary to store income sources and amounts (e.g., "salary", "freelance").
3. Create a second dictionary for expense categories and amounts (e.g., "rent", "food").
4. Write a function to calculate total income by summing dictionary values.
5. Write a function to calculate total expenses similarly.
6. Compute the balance by subtracting total expenses from total income.
7. Print a formatted summary showing all three values.
8. **ðŸ”¹ Expected Output:** Printed summary of total income, expenses, and balance (e.g., "Income: 3000, Expenses: 2000, Balance: 1000").

```python
income_sourse={"salary":50000,"freelance":18000}
expense_cate={"rent":5000,"food":4000,"cosmetics":5000,"others":6000}
```

```python
def tot_inc():
    return sum(income_sourse.values())
```

```python
def tot_exp():
    return sum(expense_cate.values())
```

```python
balance = tot_inc() - tot_exp()
print(f"Total Income:{tot_inc()} Total Expense:{tot_exp()} Balance:{balance}")
```

```python
import threading
import logging
import pandas as pd
```

```python
logging.basicConfig(filename="processing.log", level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")
```

```python
def process_file(file_name, column_name):
    logging.info(f"Started processing {file_name}")
    try:
        df = pd.read_csv(file_name)
        mean_value = df[column_name].mean()
        logging.info(f"Finished processing {file_name}) - Mean: {mean_value}")
        print(f"Mean f {column_name} in {file_name}): {mean_value:.2f}")
    except Exception as e:
        logging.error(f"Error processing {file_name}: {e}")
```

```python
file1, file2 = "/content/Titanic-Dataset.csv","/content/Titanic Dataset.csv"
column = "SibSp"
```

```python
thread1 = threading.Thread(target=process_file, args=(file1, column))
thread2 = threading.Thread(target=process_file, args=(file2, column))
```

```python
thread1.start()
thread2.start()
```

```python
thread1.join()
thread2.join()
```
