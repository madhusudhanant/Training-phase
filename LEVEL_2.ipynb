```python
my_list = [1,2,3,4,5]
```

```python
e_number = filter(lambda my_list: my_list%2==0, my_list)
even_number = list(e_number)
```

```python
c_number = map(lambda even_number: even_number**3, even_number)
cube_number = list(c_number)
print(cube_number)
```

#Bonus advanced EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/tested.csv")
df.drop(columns= ["Name", "Sex", "Ticket", "Cabin", "Embarked"], axis=1,inplace=True)
```

```python
correlation_matrix = df.corr()
heatmap = sns.heatmap(correlation_matrix, annot = True, cmap ="magma", fmt = ".2f", linewidths= 1)
plt.show(heatmap)
```
